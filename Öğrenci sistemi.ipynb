```python
import pandas as pd 
import numpy as np
```

```python
isim=[]
no=[]
sınav_notu=[]
harf_notu=[]
gecme=[]
```

```python
def sisteme_kayıt():
    isim.append(input("İsim & Soyisim giriniz. "))
    no.append(input("Okul numarası giriniz.  "))
    sınav_notu.append(int(input("Sınav notu giriniz. ")))
    
    if sınav_notu[-1]>= 90 :
        harf_notu.append("AA")
        gecme.append("Geçti")
         
            
    elif sınav_notu[-1] >= 85: 
        harf_notu.append("BA")
        gecme.append("Geçti")
         
            
    elif sınav_notu[-1] >= 80: 
        harf_notu.append("BB")
        gecme.append("Geçti")
         
    elif sınav_notu[-1] >= 75: 
        harf_notu.append("CB")
        gecme.append("Geçti")
         
    elif sınav_notu[-1] >= 70: 
        harf_notu.append("CC")
        gecme.append("Geçti")
         
    elif sınav_notu[-1] >= 65: 
        harf_notu.append("DC")
        gecme.append("Şartlı Geçti")
         
    elif sınav_notu[-1] >= 60: 
        harf_notu.append("DD")
        gecme.append("Kaldı")
         
    elif sınav_notu[-1] >= 50: 
        harf_notu.append("FD")
        gecme.append("Kaldı")
         
    elif sınav_notu[-1]<50:
        harf_notu.append("FF")
        gecme.append("Kaldı")
        
      
```
