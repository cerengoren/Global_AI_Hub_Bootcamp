```python
import pandas as pd 
import numpy as np
```

```python
isim=[]
no=[]
snav_notu=[]
harf_notu=[]
gecme=[]
```

```python
def sisteme_kayt():
    isim.append(input("sim & Soyisim giriniz. "))
    no.append(input("Okul numaras giriniz.  "))
    snav_notu.append(int(input("Snav notu giriniz. ")))
    
    if snav_notu[-1]>= 90 :
        harf_notu.append("AA")
        gecme.append("Geti")
         
            
    elif snav_notu[-1] >= 85: 
        harf_notu.append("BA")
        gecme.append("Geti")
         
            
    elif snav_notu[-1] >= 80: 
        harf_notu.append("BB")
        gecme.append("Geti")
         
    elif snav_notu[-1] >= 75: 
        harf_notu.append("CB")
        gecme.append("Geti")
         
    elif snav_notu[-1] >= 70: 
        harf_notu.append("CC")
        gecme.append("Geti")
         
    elif snav_notu[-1] >= 65: 
        harf_notu.append("DC")
        gecme.append("artl Geti")
         
    elif snav_notu[-1] >= 60: 
        harf_notu.append("DD")
        gecme.append("Kald")
         
    elif snav_notu[-1] >= 50: 
        harf_notu.append("FD")
        gecme.append("Kald")
         
    elif snav_notu[-1]<50:
        harf_notu.append("FF")
        gecme.append("Kald")
        
      
```
